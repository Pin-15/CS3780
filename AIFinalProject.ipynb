```python
import pandas as pd
url="https://raw.githubusercontent.com/shavi2006/d1/main/RetentionSep1.csv"
df=pd.read_csv(url)
df.head()
```

```python
import numpy as np
df["Yr2"]=np.where(df["Yr2"]=="X",0,df["Yr2"])
df["Yr2"]=np.where(df["Yr2"]=="Y",0,df["Yr2"])
df["Yr2"].fillna(1,inplace=True)
df["Gender"].fillna("N/A",inplace=True)
df.isnull().sum()
df["Gender"].value_counts()
df["ResdCode"].fillna("R",inplace=True)
df["Major"].value_counts()
df["Major"].fillna("N/A",inplace=True)
df["GPA1stSem"].value_counts()
df['GPA1stSem'].fillna(df['GPA1stSem'].mean(), inplace=True)
df["Admit Type"].value_counts()
df["Admit Type"].fillna("FR",inplace=True)
df["Summer PAL"].fillna("N",inplace=True)
df["Summer PAL"].value_counts()
df['HS GPA'].fillna(df['HS GPA'].median(), inplace=True)
df['First Gener'].fillna('N', inplace=True)
df['EnterTransfHrs'].fillna(df['EnterTransfHrs'].mean(), inplace=True)
df['AttHrs Yr1'].fillna(df['AttHrs Yr1'].mean(), inplace=True)
df["Earn Hrs < 30"].fillna("N", inplace=True)
df["Majr at Adm "].value_counts()
df['Majr at Adm '].fillna('NSG', inplace=True)
df['FA'].fillna('P', inplace=True)
df["Spring Census Return"].fillna("1.0",inplace=True)
df["GPA 1st Term Group"].fillna("3.50 to 4.00",inplace=True)
df["HS GPA Group"].fillna("3.00 to 3.24",inplace=True)

df['PR State'] = df['PR State'].fillna('MA')

most_repeated = df['IPEDS Race/Ethnicity'].value_counts().idxmax()
print(most_repeated)

df['IPEDS Race/Ethnicity'] = df['IPEDS Race/Ethnicity'].fillna('Race and Ethnicity Unknown')

most_repeated = df['ResHall'].value_counts().idxmax()
print(most_repeated)

df['ResHall'] = df['ResHall'].fillna('COMMUTER')

most_repeated = df['ResdCode'].value_counts().idxmax()
print(most_repeated)

df['SO'] = df['SO'].fillna("N")
df.isnull().sum()
import seaborn as sns
sns.histplot(df["RegTermCred for Yr1 Fall"])
dm=df["RegTermCred for Yr1 Fall"].mean()
df["RegTermCred for Yr1 Fall"].fillna(dm,inplace=True)
sns.histplot(df["RegTermCred for Yr1 Spring"])
le=df["RegTermCred for Yr1 Spring"].mean()
df["RegTermCred for Yr1 Spring"].fillna(le,inplace=True)
df["Sports sum"].fillna("0.0",inplace=True)

df["AcdStand1stSem"].fillna("00",inplace=True)

df.isnull().sum()
df["HS GPA Group"].fillna("3.00 to 3.24",inplace=True)
df["GPA 1st Term Group"].value_counts()
df.dropna(subset=["GPA 1st Term Group"], inplace=True)

df["FA"].fillna("N/A", inplace=True)
```

```python
df["Yr2"]=df["Yr2"].astype("str")
df["Gender"]=df["Gender"].astype("str")
df["ResdCode"]=df["ResdCode"].astype("str")
df["Major"]=df["Major"].astype("str")
df["Admit Type"]=df["Admit Type"].astype("str")
df["Summer PAL"]=df["Summer PAL"].astype("str")
df["Majr at Adm "]=df["Majr at Adm "].astype("str")
df["FA"]=df["FA"].astype("str")
df["Spring Census Return"]=df["Spring Census Return"].astype("str")
df["GPA 1st Term Group"]=df["GPA 1st Term Group"].astype("str")
df["HS GPA Group"]=df["HS GPA Group"].astype("str")
df["Earn Hrs < 30"]=df["Earn Hrs < 30"].astype("str")
df["AttHrs Yr1"]=df["AttHrs Yr1"].astype("int")
df["EnterTransfHrs"]=df["EnterTransfHrs"].astype("int")
df["First Gener"]=df["First Gener"].astype("str")


```

```python
df.head()
```

```python
df=pd.get_dummies(df, drop_first=True)
df.head()

```

```python
df["GPA1stSem"] = (df["GPA1stSem"] - df["GPA1stSem"].min()) / (df["GPA1stSem"].max() - df["GPA1stSem"].min())
df["HS GPA"] = (df["HS GPA"] - df["HS GPA"].min()) / (df["HS GPA"].max() - df["HS GPA"].min())
df["EnterTransfHrs"] = (df["EnterTransfHrs"] - df["EnterTransfHrs"].min()) / (df["EnterTransfHrs"].max() - df["EnterTransfHrs"].min())
df["AttHrs Yr1"] = (df["AttHrs Yr1"] - df["AttHrs Yr1"].min()) / (df["AttHrs Yr1"].max() - df["AttHrs Yr1"].min())
df["RegTermCred for Yr1 Fall"] = (df["RegTermCred for Yr1 Fall"] - df["RegTermCred for Yr1 Fall"].min()) / (df["RegTermCred for Yr1 Fall"].max() - df["RegTermCred for Yr1 Fall"].min())
df["RegTermCred for Yr1 Spring"] = (df["RegTermCred for Yr1 Spring"] - df["RegTermCred for Yr1 Spring"].min()) / (df["RegTermCred for Yr1 Spring"].max() - df["RegTermCred for Yr1 Spring"].min())
df.describe()
```

```python
before=df.iloc[:,:6]
after=df.iloc[:,7:]
target=df.iloc[:,6]
df=pd.concat([before,after,target], axis=1,join="outer")
print(list(df.columns))
```

```python
y = df.iloc[:, -1]
X = df.iloc[:, :-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
